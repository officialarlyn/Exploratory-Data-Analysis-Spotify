```python
import  pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify_top_songs_audio_features.csv')
```

```python
df
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop any duplicate values
df.drop_duplicates()
```

```python
# Converting the duration_ms to minutes

df['duration_min'] = (df['duration_ms']/(1000*60)).round(2)
```

```python
df
```

```python
# dropping the duration_ms column
df.drop('duration_ms', axis=1, inplace=True)
```

```python
df.head()
```

```python

# Detecting Outliers using the IQR for the tempo column
df['tempo'] = df['tempo'].astype(int)

Q1 = df['tempo'].quantile(0.25)
Q3 = df['tempo'].quantile(0.75)
IQR = Q3 - Q1

# Defining the upper & lower bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset for outliers
lower_outliers = df['tempo'] < lower_bound
upper_outliers = df['tempo'] > upper_bound

tempo_outliers = lower_outliers | upper_outliers
```

```python
# Dropping the outliers in tempo
df = df[~tempo_outliers]
```

```python
# Detecting Outliers using IQR for  the streams column
Q1 = df['streams'].quantile(0.25)
Q3 = df['streams'].quantile(0.75)
IQR = Q3 - Q1

# Defining the lower & upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
lower_outliers = df['streams'] < lower_bound
upper_outliers = df['streams'] > upper_bound

# Filter the dataset for outliers
streams_outliers = lower_outliers | upper_outliers
```

```python
# Dropping the Outliers in streams
df = df[~streams_outliers]
```

```python
df
```

```python
# Check the descriptive statistics of the various numerical columns
df.describe()
```

```python
# Correlation Analysis
numeric_data = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_data.corr()
corr_matrix
```

```python
# Heatmap for correlation analysis
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Show the distribution of song durations in minutes
sns.histplot(df['duration_min'], kde=True, bins=30)  # Added kde for a smoother curve
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for popularity
plt.subplot(1, 2, 1)
sns.histplot(df['streams'], bins=20, kde=True)
plt.title('Distribution of streams')
```

```python
# Scatter plot for streams vs. danceability
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['danceability'], y=df['streams'])
plt.title('streams vs Danceability')
plt.xlabel('Danceability')
plt.ylabel('streams')
plt.show()
```

```python
# Top artists by frequency
top_artists = df['artist_names'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.index, y=top_artists.values)
plt.title('Top 10 Artists by Frequency')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```
